```python
import random

def exibir_tabuleiro(tabuleiro):
    """Imprime o tabuleiro no formato atual."""
    print("+-------+-------+-------+")
    for linha in tabuleiro:
        print("|       |       |       |")
        print(f"|   {linha[0]}   |   {linha[1]}   |   {linha[2]}   |")
        print("|       |       |       |")
        print("+-------+-------+-------+")

def campos_livres(tabuleiro):
    """Retorna uma lista de tuplas (linha, coluna) de quadrados vazios."""
    livres = []
    for r in range(3):
        for c in range(3):
            if tabuleiro[r][c] not in ['O', 'X']:
                livres.append((r, c))
    return livres

def movimento_usuario(tabuleiro):
    """Solicita e valida o movimento do jogador."""
    while True:
        try:
            escolha = int(input("Escolha seu movimento (1-9): "))
            if escolha < 1 or escolha > 9:
                print("Número inválido! Escolha entre 1 e 9.")
                continue

            # Converte o número 1-9 para índices de matriz (linha, coluna)
            linha = (escolha - 1) // 3
            coluna = (escolha - 1) % 3

            if (linha, coluna) in campos_livres(tabuleiro):
                tabuleiro[linha][coluna] = 'O'
                break
            else:
                print("Esse campo já está ocupado!")
        except ValueError:
            print("Entrada inválida! Digite um número inteiro.")

def movimento_computador(tabuleiro):
    """O computador escolhe um campo aleatório entre os disponíveis."""
    livres = campos_livres(tabuleiro)
    if livres:
        # Escolha aleatória simples
        r, c = random.choice(livres)
        tabuleiro[r][c] = 'X'

def verificar_vitoria(tabuleiro, sinal):
    """Verifica se o jogador com o 'sinal' (X ou O) venceu."""
    # Verificar linhas e colunas
    for i in range(3):
        if all(tabuleiro[i][j] == sinal for j in range(3)): return True
        if all(tabuleiro[j][i] == sinal for j in range(3)): return True

    # Verificar diagonais
    if all(tabuleiro[i][i] == sinal for i in range(3)): return True
    if all(tabuleiro[i][2-i] == sinal for i in range(3)): return True

    return False

# --- Fluxo Principal do Jogo ---

# Inicializa o tabuleiro com os números de 1 a 9
quadro = [[str(3 * j + i + 1) for i in range(3)] for j in range(3)]

# Regra: O computador sempre começa no meio (posição 5, que é índice [1][1])
quadro[1][1] = 'X'

while True:
    exibir_tabuleiro(quadro)

    # Turno do Usuário
    movimento_usuario(quadro)
    if verificar_vitoria(quadro, 'O'):
        exibir_tabuleiro(quadro)
        print("Você ganhou! Parabéns!")
        break

    if not campos_livres(quadro):
        exibir_tabuleiro(quadro)
        print("Empate!")
        break

    # Turno do Computador
    movimento_computador(quadro)
    if verificar_vitoria(quadro, 'X'):
        exibir_tabuleiro(quadro)
        print("O computador ganhou! Mais sorte na próxima.")
        break

    if not campos_livres(quadro):
        exibir_tabuleiro(quadro)
        print("Empate!")
        break
```
